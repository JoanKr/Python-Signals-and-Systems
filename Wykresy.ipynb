```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import sympy 
default_figsize=(14, 8)
test_vec =  np.array([-1, 1, 0, 4, -3, -6, 7, 9, 1, 4, -9, -5])
test_poly = np.poly1d([1, 2, -5])
```

```python
def introduction_plot_sinus():
    #narysuj wykres    
    x = np.linspace(0, 2*np.pi, 2000) #przedzial w ktorym rysujemy
    y = np.sin(x+np.pi/4) #wzor sinusa
    fig = plt.figure(figsize = default_figsize)
    plt.plot(x, y, "m")

    #dodaj opisy osi i tytuł wykresu
    plt.title("sinus")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid()
        
    #zastosować 10% marginesy
    x_min = np.min(x)
    x_max = np.max(x)
    x_range = x_max - x_min
    x_margin = x_range / 10

    y_min = np.min(y)
    y_max = np.max(y)
    y_range = y_max - y_min
    y_margin = y_range / 10

    plt.axis([x_min - x_margin, x_max + x_margin, y_min - y_margin, y_max + y_margin])
    
introduction_plot_sinus()
```

```python
def introduction_plot_subplot():
    #narysuj wykres
    x = np.linspace(0, 2*np.pi, 2000)
    y = np.sin(x)
    z = np.cos(x)
    w = np.arctan(x)

    fig = plt.figure(figsize = default_figsize)

    #sinus
    plt.subplot(2,1,1)
    plt.plot(x, y, "b")
    plt.grid()

    #cosinus
    plt.subplot(2,2,3)
    plt.plot(x, z , "r")
    plt.grid()

    #arctg
    plt.subplot(2,2,4)
    plt.plot(x, w, "m")
    plt.grid()

introduction_plot_subplot()
```

```python
def introduction_plot_polynomial():
    #wykonaj operacje na wielomianach
    a = np.poly1d([1, 1, -1, 0, -1])
    b = np.poly1d([1, 0, -1, 1])
    c = np.poly1d([1, -1, 1, 0, 0, -1])
    d = np.polymul(a,c)
    e = np.polyder(d)
    f = e + b - np.poly1d([1,0,0,0])

    #wyświetl wielomiany
    print("a=" + str(a.coeffs))
    print("b=" + str(b.coeffs))
    print("c=" + str(c.coeffs))
    print("d=" + str(d.coeffs))
    print("e=" + str(e.coeffs))
    print("f=" + str(f.coeffs))
    print()

    #narysuj wykres
    x = np.linspace(-1, 1, 2000)
    y = np.polyval(f, x)
    fig = plt.figure(figsize = default_figsize)
    plt.plot(x, y, "r")
    plt.grid()

introduction_plot_polynomial()
```

```python
def introduction_symbolic():
    #zadeklaruj zmienną t
    t = sympy.Symbol('t')

    #wyznacz i wyświetl całkę nieoznaczoną
    display(sympy.integrate(t**2,t))
    print()
    #wyznacz i wyświetl całkę oznaczoną
    display(sympy.integrate(t**2, (t, -1, 1)))

introduction_symbolic()
```

```python
def introduction_plot_histogram():
    fig = plt.figure(figsize = default_figsize)
    #wygeneruj wektory elementów
    a = np.random.uniform(-1,1,100000)
    b = np.random.normal(0, 0.5, 100000)

    #narysuj histogramy
    plt.subplot(1,2,1)
    plt.hist(a, bins=50, color="red")
    plt.grid()
    plt.subplot(1,2,2)
    plt.hist(b, bins=50, color="blue")
    plt.grid()   
    
introduction_plot_histogram()
```

```python
def taskA(y):
    num_samples=len(y)
    x = np.arange(0, num_samples)
    fig = plt.figure(figsize = default_figsize)
    plt.margins(0,1)
    plt.plot(x, y, "o")
    #plt.scatter(x, y)
    plt.grid()

taskA(test_vec)
```

```python
#funkcja sortująca wektor
import copy
def someKindOfSort(src):
    data = copy.copy(src)
    n = len(data)
    for i in range(n):
        for j in range(0, n-1):
            if abs(data[j])>abs(data[j+1]):
                data[j], data[j+1] = data[j+1], data[j]
    return data

def taskB(data2):
    resultSorted = someKindOfSort(data2)
    print(resultSorted)

taskB(test_vec)
```

```python
#funkcja wyznaczająca pierwiastki
import math
def taskC(polynom):
    #a*x^2 + b*x + c
    #test_poly = np.poly1d([1, 2, -5])
    #d < 0 print ("brak pierwiastkow rzeczywistych")

    a = polynom[2]
    b = polynom[1]
    c = polynom[0]

    delta = b**2 - 4*a*c
    if (delta > 0):
        x1 = ((-b) - math.sqrt(delta))/(2*a)
        x2 = ((-b) + math.sqrt(delta))/(2*a)
        print ("roots --> x1={}, x2={}".format(x1, x2))
        print()
    elif(delta==0):
        x12= (-b)/(2*a)
        print("roots --> x1=x2={}".format(x12))
        print()
    else:
        print("roots --> no real roots")
        print()

    x = np.linspace(-4, 4, 2000)
    y = np.polyval(polynom, x)
    fig = plt.figure(figsize = default_figsize)
    plt.plot(x, y, "r")
    plt.grid()

taskC(test_poly)
```

```python
#przetestuj funkcję o argumentach będących współczynnikami wielomianu kwadratowego oraz zakresem i dokładnością punktów dziedziny, wykreślającą ten wielomian w podanym zakresie
def taskD(polynom, xrange, num_points):
    figure = plt.figure(figsize = default_figsize)
    x = np.linspace(xrange[0], xrange[1], num_points)
    y = np.polyval(polynom, x)
    plt.plot(x ,y , "m")
    plt.grid()

taskD(test_poly, [-6, 4], 400)
```
